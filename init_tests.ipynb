```python
from graphviz import Digraph
import torch.nn as nn

def visualize_tensordict_module(td_module, filename='tensordict_module'):
    dot = Digraph(comment='TensorDictModule Visualization')
    
    # Assuming td_module.module is a nn.Sequential for simplicity
    sequential_module = td_module.module
    prev_node_name = "Input"
    dot.node(prev_node_name, label=f"Input\nKeys: {td_module.in_keys}")
    
    for i, module in enumerate(sequential_module):
        module_name = f"{i}: {module.__class__.__name__}"
        input_shape = "Unknown"  # Determining input/output shapes dynamically can be tricky
        output_shape = "Unknown"
        
        # Customize this part to fetch actual shapes if possible
        if isinstance(module, nn.Linear):
            input_shape = module.in_features
            output_shape = module.out_features
        
        label = f"{module_name}\nIn: {input_shape}\nOut: {output_shape}"
        dot.node(module_name, label=label)
        dot.edge(prev_node_name, module_name)
        prev_node_name = module_name
    
    # Adding the final output node
    dot.node("Output", label=f"Output\nKeys: {td_module.out_keys}")
    dot.edge(prev_node_name, "Output")
    
    # Render the graph to a file (e.g., PDF, PNG)
    dot.render(filename, format='png', cleanup=True)
    print(f"Graph is saved as {filename}.png")
```

```python
from graphviz import Digraph
import torch.nn as nn

def visualize_single_nn_model_with_subgraph(graph, model, in_key="input", out_key="output"):
    # Use HTML-like labels for more control over formatting
    subgraph_label = f'<<B>Neural Network Layers</B><BR/>Input key: <FONT COLOR="blue">{in_key}</FONT><BR/>Output key: <FONT COLOR="blue">{out_key}</FONT>>'

    with graph.subgraph(name='cluster_nn_layers') as c:
        c.attr(color='blue', label=subgraph_label)
        c.node_attr.update(style='filled', fillcolor='lightgrey')

        prev_name = None
        for i, layer in enumerate(model):
            layer_name = f'{type(layer).__name__}{i}'
            label = f'{type(layer).__name__}'
            if isinstance(layer, nn.Linear):
                label += f'\nIn: {layer.in_features}, Out: {layer.out_features}'
            elif isinstance(layer, nn.Dropout):
                label += f'\nP: {layer.p}'
            
            c.node(layer_name, label, shape='box')

            if prev_name is not None:
                c.edge(prev_name, layer_name)
            prev_name = layer_name

# Define your model
model = nn.Sequential(nn.Linear(4, 5), nn.ReLU(), nn.Dropout(p=0.1), nn.Linear(5, 2))

graph = Digraph()

# Visualize it
visualize_single_nn_model_with_subgraph(graph, model, in_key="input", out_key="output")
# Render and view
graph.view()
```

```python
class TensorDictVizGraph:
    def __init__(
        self,
        model: Optional[
            Union[nn.Module, nn.Sequential, TensorDictModule, TensorDictSequential]
        ] = None,
        filename: str = "tensordict_module",
    ):
        self.graph = Digraph(comment="TensorDictModule Visualization")
        self.filename = filename
        self.model = model

    def build_graph(
        self,
        model: Optional[
            Union[nn.Module, nn.Sequential, TensorDictModule, TensorDictSequential]
        ] = None,
    ):
        if model is None:
            model = self.model
        if model is None and self.model is None:
            raise ValueError("Please provide a model to visualize")
        if isinstance(model, TensorDictModule):
            self.td_module_subgraph(model)
        elif isinstance(model, tensordict.nn.TensorDictSequential):
            self.td_sequential_graph(model)
        elif isinstance(model, nn.Sequential):
            self.nn_module_subgraph(model)
        elif isinstance(model, nn.Module):
            self.nn_module_subgraph(model.model)


    
            
    def td_sequential_graph(self, sequential_td_module):
        self.graph.attr(compound="true", rankdir="TB")
        self.graph.node("start", "Start", shape="Mdiamond")
        graph_dict = {}

        prev_out_key = None
        prev_module_name = None
        prev_out_edge = "start"

        for i, td_module in enumerate(sequential_td_module):
            module_name = f"module_{i}"
            in_key = ", ".join(td_module.in_keys)
            out_key = ", ".join(td_module.out_keys)
            out_edge = self.nn_module_subgraph(td_module.module, module_name, in_key, out_key, return_final=True)
            graph_dict[module_name] = (in_key, out_key, out_edge)

            # Connect the start node to the first module
            if i == 0:
                self.graph.edge("start", f"{module_name}:{in_key}", style="bold")

            # Connect this module's subgraph to the previous one if the out_key matches the in_key
            if prev_out_key and any(k in prev_out_key for k in in_key.split(", ")):
                self.graph.edge(f"{prev_module_name}:{prev_out_key}", f"{module_name}:{in_key}", style="bold")

            prev_out_key = out_key
            prev_module_name = module_name
            prev_out_edge = out_edge

        # This print statement is for debugging; you can remove it once you confirm the graph is built correctly
        print(graph_dict)

    def td_module_subgraph(self, td_module, subgraph_name:str="cluster_nn_layers",):
        self.nn_module_subgraph(
            td_module.module, subgraph_name, in_key=td_module.in_keys, out_key=td_module.out_keys
        )

    def nn_module_subgraph(
        self, model: nn.Sequential, subgraph_name:str="cluster_nn_layers", in_key="input", out_key="output", return_final=False
    ):
        # Use HTML-like labels for more control over formatting
        subgraph_label = f'<<B>Neural Network Layers</B><BR/>Input key: <FONT COLOR="blue">{in_key}</FONT><BR/>Output key: <FONT COLOR="blue">{out_key}</FONT>>'

        with self.graph.subgraph(name=subgraph_name) as c:
            c.attr(color="blue", label=subgraph_label)
            c.node_attr.update(style="filled", fillcolor="lightgrey")

            prev_name = None
            for i, layer in enumerate(model):
                layer_name = f"{type(layer).__name__}{i}"
                label = f"{type(layer).__name__}"
                if isinstance(layer, nn.Linear):
                    label += f"\nIn: {layer.in_features}, Out: {layer.out_features}"
                elif isinstance(layer, nn.Dropout):
                    label += f"\nP: {layer.p}"
                elif isinstance(layer, nn.Conv2d):
                    label += f"\nIn Channel: {layer.in_channels}, Out Channel: {layer.out_channels}"
                    label += f"\nKernel: {layer.kernel_size}, Stride: {layer.stride}"

                c.node(layer_name, label, shape="box")

                if prev_name is not None:
                    c.edge(prev_name, layer_name)
                prev_name = layer_name
        if return_final:
            return prev_name
    def view(
        self,
    ):
        self.graph.view()

    def clear(
        self,
    ):
        self.graph.clear()

    def render(
        self,
    ):
        self.graph.render(self.filename, format="png", cleanup=True)
        print(f"Graph is saved as {self.filename}.png")
```

```python
# Example activation class
activation_class = nn.ReLU  # You can replace nn.ReLU with any other activation class e.g., nn.Tanh

class ConvNet(nn.Module):
    def __init__(self, activation_class, num_cells, kernel_sizes, strides):
        super(ConvNet, self).__init__()
        
        layers = []
        in_channels = 1  # Assuming the input has 1 channel, adjust accordingly for your dataset
        
        # Construct the convolutional layers based on the provided parameters
        for out_channels, kernel_size, stride in zip(num_cells, kernel_sizes, strides):
            layers.append(nn.Conv2d(in_channels, out_channels, kernel_size, stride))
            layers.append(activation_class())
            in_channels = out_channels  # Update in_channels for the next layer
        
        # Combine all layers into a sequential model
        self.model = nn.Sequential(*layers)
    
    def forward(self, x):
        return self.model(x)

# Example usage
cnn = ConvNet(
    activation_class=activation_class,
    num_cells=[32, 64, 64],
    kernel_sizes=[8, 4, 3],
    strides=[4, 2, 1],
)
```

```python
class TensorDictVizGraph:
    def __init__(
        self,
        model: Optional[
            Union[nn.Module, nn.Sequential, TensorDictModule, TensorDictSequential]
        ] = None,
        filename: str = "tensordict_module",
    ):
        self.graph = Digraph(comment="TensorDictModule Visualization")
        self.filename = filename
        self.model = model

    def build_graph(
        self,
        model: Optional[
            Union[nn.Module, nn.Sequential, TensorDictModule, TensorDictSequential]
        ] = None,
    ):
        if model is None:
            model = self.model
        if model is None and self.model is None:
            raise ValueError("Please provide a model to visualize")
        if isinstance(model, TensorDictModule):
            self.td_sequential_graph(model)
        elif isinstance(model, tensordict.nn.TensorDictSequential):
            self.td_sequential_graph(model)
        elif isinstance(model, nn.Sequential):
            self.nn_module_subgraph(model, "cluster_nn_layers", "input", "output")
        elif isinstance(model, nn.Module):
            self.nn_module_subgraph(model.model, "cluster_nn_layers", "input", "output")

    def td_sequential_graph(self, sequential_td_module):
        self.graph.attr(compound="true", rankdir="TB")
        self.graph.node("start", "Input TensorDict", shape="box")

        prev_module_exit = (
            "start"  # Initialize the connection start point as the "Start" node.
        )
        prev_in_key = None
        prev_out_key = None
        graph_dict = {}
        for i, td_module in enumerate(sequential_td_module):
            module_name = "cluster" + f"module_{i}"
            in_key = ", ".join(td_module.in_keys)
            out_key = ", ".join(td_module.out_keys)
            in_key = td_module.in_keys
            out_key = td_module.out_keys

            entry_node, exit_node = self.nn_module_subgraph(
                td_module.module, module_name, in_key, out_key
            )
            if i == 0:
                graph_dict[module_name] = {
                    "in_key": in_key,
                    "out_key": out_key,
                    "entry_node": entry_node,
                    "exit_node": exit_node,
                    "prev_module_exit": prev_module_exit,
                }
            else:
                if prev_out_key == in_key:
                    graph_dict[module_name] = {
                        "in_key": in_key,
                        "out_key": out_key,
                        "entry_node": entry_node,
                        "exit_node": exit_node,
                        "prev_module_exit": prev_module_exit,
                    }
                else:
                    # search for the previous module exit
                    found_prev_module_exit = None
                    for k, v in graph_dict.items():
                        if v["out_key"] == in_key:
                            assert found_prev_module_exit is None, "Multiple modules found with the same out_key"
                            found_prev_module_exit = graph_dict[k]["exit_node"]
                        graph_dict[module_name] = {
                            "in_key": in_key,
                            "out_key": out_key,
                            "entry_node": entry_node,
                            "exit_node": exit_node,
                            "prev_module_exit": found_prev_module_exit,
                        }
                        break
                    # if its still none then it must be connected to start
                    if found_prev_module_exit is None:
                        prev_module_exit = "start"
                        graph_dict[module_name] = {
                            "in_key": in_key,
                            "out_key": out_key,
                            "entry_node": entry_node,
                            "exit_node": exit_node,
                            "prev_module_exit": prev_module_exit,
                        }

            prev_in_key = in_key
            prev_out_key = out_key
            prev_module_exit = (
                exit_node  # Update the exit point for the next iteration.
            )

        print(graph_dict)

        # draw the edges
        for k, v in graph_dict.items():
            arrow_start = v["prev_module_exit"]
            arrow_end = v["entry_node"]
            
            self.graph.edge(arrow_start, arrow_end, style="solid")

        # end node
        #out_keys = []
        #self.graph.node("end", "Output TensorDict\n+outkeys", shape="box")

    def nn_module_subgraph(
        self, model, subgraph_name, entry_node_label, exit_node_label
    ):
        # Use HTML-like labels for more control over formatting
        subgraph_label = f'<<B>Input key:</B> <FONT COLOR="blue">{entry_node_label}</FONT><BR/><B>Output key:</B> <FONT COLOR="blue">{exit_node_label}</FONT>>'

        with self.graph.subgraph(name=subgraph_name) as c:
            c.attr(color="blue", label=subgraph_label)
            c.node_attr.update(style="filled", fillcolor="lightgrey")

            # Define entry and exit points for the subgraph
            entry_node = f"{subgraph_name}_entry"
            exit_node = f"{subgraph_name}_exit"
            c.node(
                entry_node,
                label=f"Entry: {entry_node_label}",
                shape="point",
                style="invis",
            )
            prev_name = entry_node  # Start connecting from the entry node

            for i, layer in enumerate(model):
                layer_name = f"{subgraph_name}_{type(layer).__name__}{i}"
                label = f"{type(layer).__name__}"
                if isinstance(layer, nn.Linear):
                    label += f"\nIn: {layer.in_features}, Out: {layer.out_features}"
                elif isinstance(layer, nn.Dropout):
                    label += f"\nP: {layer.p}"
                elif isinstance(layer, nn.Conv2d):
                    label += f"\nIn: {layer.in_channels}, Out: {layer.out_channels}"
                    label += f"\nKernel: {layer.kernel_size}, Stride: {layer.stride}"

                c.node(layer_name, label, shape="box")
                c.edge(prev_name, layer_name)
                prev_name = layer_name  # Update the prev_name for the next iteration

            # Connect the last layer to the exit node and make the exit node invisible
            c.node(
                exit_node,
                label=f"Exit: {exit_node_label}",
                shape="point",
                style="invis",
            )
            c.edge(prev_name, exit_node)

        # Returning the entry and exit node labels for potential external connections
        return entry_node, exit_node

    def view(
        self,
    ):
        self.graph.view()

    def clear(
        self,
    ):
        self.graph.clear()

    def render(
        self,
    ):
        self.graph.render(self.filename, format="png", cleanup=True)
        print(f"Graph is saved as {self.filename}.png")
```
